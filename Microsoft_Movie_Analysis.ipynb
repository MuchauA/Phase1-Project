**2. Performing data cleaning on bom.movie_gross DataSet**

```python
df_gross.isna().sum() #checking for the total missing values in the bom_movie_gross dataset
```

*Check the datatypeof each column*

```python
df_gross.info()
```

```python
df_gross['foreign_gross'].dtype
```

***convert the DataType of the column "foreign_gross" from object type to float***

```python

df_gross['foreign_gross'] = pd.to_numeric(df_gross['foreign_gross'], errors = 'coerce')
df_gross
```

**assert the dtype has changed***

```python
df_gross['foreign_gross'].dtype
```

```python
df_gross.isna().sum() #look for missing values remaining in the movie_gross dataset
```

*From the bom.movie_gross, the foreign_gross column has 1350 null values. Since the values are numerical, we replace the missing values with the column median*

```python
df_gross['foreign_gross'] = df_gross['foreign_gross'].fillna(df_gross['foreign_gross'].median())
df_gross
```

```python
df_gross.isna().sum() #check for total null values, per column,  in the movie_gross dataset
```

```python
df_gross['domestic_gross'].dtype #finding the datatype of the domestic_gross column
```

*domestic_gross column has 28 missing values. We replace the missing values with the median*

```python
df_gross['domestic_gross'] = df_gross['domestic_gross'].fillna(df_gross['domestic_gross'].median())
df_gross
```

```python
df_gross['domestic_gross'].median() #finding the median of the domestic_gross column
```

*ascertain there are no null values in domestic_gross column*

```python
df_gross['domestic_gross'].isna().sum()
```

*Studio column has 5 null values,we replace them with the most common value in the column*

```python
df_gross['studio'].value_counts() #check for the total number of values for each unique entry in the studio column of bom.movie_gross dataset
```

*The most common value in studio column is IFC, hence, we replace the null values with IFC*

```python
df_gross['studio'] = df_gross['studio'].fillna('IFC')
df_gross
```

Assert the values are filled and there are no missing values in the dataset

```python
df_gross.isna().sum()
```
